```r
library(clusterProfiler)
library(tidyverse)
```

```r
rna_hichip = read.csv('../data/processed/fig1/rna_hichip_df.csv')
rna_hichip = rna_hichip %>%
    mutate(intersect_over_min = if_else(pval_bonf<0.05, intersect_over_min, 0))%>%
    mutate(jaccard = if_else(pval_bonf<0.05, jaccard, 0))
```

```r
ggplot(rna_hichip)+
    geom_tile(aes(x=color_col, y=color_row, fill=log_pval_bonf))+
scale_fill_gradient(low="white", high="red") +
     geom_count(data=rna_hichip, aes(x=color_col, y=color_row,size=jaccard),colour="black",pch=21)+
 scale_size_area()+
  scale_size_area(max_size = 20)+
    theme_minimal()
ggsave('../data/processed/fig1/rna_hichip_df.pdf')
# , size=intersect_over_min
```
