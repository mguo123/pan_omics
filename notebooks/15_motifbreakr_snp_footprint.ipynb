```python
snp_to_motif_dict = {'colon':'Colon','gm':"GM12878",'kcd6':'GDSD6','scc':'A431-CTRLi'}
```

# atac

nothing was in footprint

```python
bed_snp_regions = pd.DataFrame()
bed_files = '../data/interim/merged/atac/'
for snp_tissue in snp_to_motif_dict:
    print(snp_tissue)
    foot_bed = pybedtools.BedTool(os.path.join(bed_files, snp_to_motif_dict[snp_tissue])+"_merged.bed")
    snp_motif_df_tissue = snp_motif_df[snp_motif_df.tissue==snp_tissue]
    print(snp_motif_df_tissue.shape)
    snp_motif_df_bed = pybedtools.BedTool.from_dataframe(snp_motif_df_tissue[['chr', 'start','stop','rsID']])
    intersect_bed = snp_motif_df_bed.intersect(foot_bed, wa=True, wb=True)
    if len(intersect_bed)>0:
        df = intersect_bed.to_dataframe().drop_duplicates()
        df['tissue'] = snp_tissue
        display(df)
        bed_snp_regions = pd.concat([bed_snp_regions, df])
                        
```

```python
bed_snp_regions.shape
```

```python
bed_snp_regions.to_csv(os.path.join(save_dir, 'bed_snp_regions_atac.csv'))
```
