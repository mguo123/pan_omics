```python
import pandas as pd
import os, glob
```

```python
all_eval = pd.DataFrame()
for idx,file in enumerate(glob.glob('../data/processed/fig4_modelling/*eval_df.csv')):
    eval_type = os.path.basename(file).split('_eval_df.')[0]
    print(eval_type)
    df = pd.read_csv(file,index_col=0)
    df.columns = ['perc', eval_type]
    if idx!=0:
        df = df[[eval_type]]
    all_eval = pd.concat([all_eval, df],axis=1)
```

```python
all_eval.to_csv('../data/processed/fig4_modelling/all_evals.csv')
```
